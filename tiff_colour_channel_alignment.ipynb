# Code for colour channel alignment of .tiff images

This code can be used for the alignment of colour channels of .tiff images. For it to work, the user should be in the Pylake environment (https://lumicks-pylake.readthedocs.io/en/stable/install.html).

```python
import lumicks.pylake as lk
import matplotlib.pyplot as plt
import os
import glob
```

```python
directory_path = r""
file_paths = glob.glob(os.path.join(directory_path, "*.tiff"))
for file in file_paths:
    stack = lk.ImageStack(file, align=True)
    directory = os.path.dirname(file)
    current_basename, file_extension = os.path.splitext(os.path.basename(file))
    text_to_add = "_aligned"
    new_basename = current_basename
    new_basename += text_to_add
    new_file_path = os.path.join(directory, new_basename + file_extension)
    stack.export_tiff(new_file_path)
```
